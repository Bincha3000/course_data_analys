```python
def exchange_rates(currency, format='full'):
    response = requests.get('https://www.cbr-xml-daily.ru/daily_json.js').json()['Valute']
    data = response[currency]
    if format == 'full':
        return data 
    elif format == 'value':
        return data['Value']
    else:
        return 'Unkown format!'
```

```python
exchange_rates(currency='USD')
```

```python
exchange_rates('EUR', 'value')
```

```python
def valute_name(id_val):
    responce = requests.get('https://www.cbr-xml-daily.ru/daily_json.js').json()['Valute']
    for key in responce.keys():
        data = responce[key]
        if id_val in data.values():
            return data['Name']
    return 'Unkown ID!'
```

```python
valute_name('R01565')
```
