```python
import requests 
from yaml import load 
from pprint import pprint 
f = open('config.yaml', 'r')
config = load(f)
token = config['token']
API_URL = 'https://api-metrika.yandex.ru/stat/v1/data'
startDate = '2018-02-12'
endDate = '2018-02-18'
counter = '21075004'
dimensions = ['ym:s:date', 'ym:s:deviceCategory', 'ym:s:browser']
metrics = ['ym:s:visits', 'ym:s:users', 'ym:s:pageviews']
filters = ['ym:s:trafficSource', 'ym:s:hasAdBlocker']
```

```python
def make_request(**kwargs):
    print(kwargs)
make_request(date1='2018-02-12', date2='2018-02-18')
```

```python
def make_request(**kwargs):
    return requests.get(API_URL, params = kwargs).json()
data = make_request(date1=startDate, date2=endDate, dimensions=','.join(dimensions),metrics=','.join(metrics),filters=','.join(filters), id = counter, oauth_token = token, accuracy = 1)
```

```python
report = []
for line in data['data']:
    dimensions_list = [x['name'] for x in line['dimensions']]
    meteics_list = line['metrics']
    report.append(dimensions_list + meteics_list)
    print(report)
    break
```

```python
def json_handler(data):
    report = []
    for line in data['data']:
        dimensions_list = [x['name'] for x in line['dimensions']]
        metrics_list = line['metrics']
        report.append(dimensions_list + metrics_list)
    return report
```

```python
json_handler(data)
```
