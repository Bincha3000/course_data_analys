```python
import pandas as pd
from datetime import datetime, timedelta
import time
data = pd.read_csv('data.tsv', sep = '\t')
data.head()
```

```python
len(data)
```

```python
def convert_to_datatime(row):
    return datetime.strptime(row['date'], '%d.%m.%Y %H:%M')
data['datetime'] = data.apply(convert_to_datatime, axis = 1)
data.head()
```

```python
def make_unix_time(row):
    return time.mktime(row['datetime'].timetuple())
data['unixtime'] = data.apply(make_unix_time, axis = 1)
data.head()
```

```python
data = data.groupby('user_id').agg(['min','max'])['unixtime']
```

```python
data.head()
```

```python
data['dif'] = data['max'] - data['min']
```

```python
data = data[data['dif'] != 0]
sec = data['dif'].mean()
```

```python
life_time = timedelta(seconds = sec)
```
