```python
import pandas as pd
import requests
from yaml import load
data = pd.read_csv('arpu_revenue_data.csv', sep = '\t')
data.head()
```

```python
f = open('config.yaml', 'r')
config = load(f)
token = config['token']
```

```python
API_URL = 'https://api-metrika.yandex.ru/stat/v1/data'
counter = '21075004'
dimensions = ['ym:s:date', 'ym:s:deviceCategory', 'ym:s:mobilePhone', 'ym:s:mobilePhoneModel']
metrics = ['ym:s:users']
metrics_string = ','.join(metrics)
startData = '2018-03-01'
endData = '2018-03-30'
dimensions_string = ','.join(dimensions)
```

```python
def report_for_offset_value(offset, n_rows):
    params = {
        'id': counter,
        'dimensions': dimensions_string,
        'metrics': metrics_string,
        'date1': startDate,
        'date2': endDate,
        'oauth_token': token,
        'offset': offset,
        'limit': n_rows
    }
    r = requests.get(API_URL, params=params)
    return r.json()
```

```python
n_rows = 10000
offset = 1
data1 = report_for_offset_value(offset, n_rows)
```

```python
len(data1['data'])
```

```python
def json_handler(data1):
    """Преобразование ответа из JSON в список листов"""
    report = []
    for line in data1['data']:
        dimensions_list = [x['name'] for x in line['dimensions']]
        metrics_list = line['metrics']
        report.append( dimensions_list + metrics_list )    
    return report
```

```python
data1 = json_handler(data1)
```

```python
dataF = {}
dataF['date'] = []
dataF['deviceCategory'] = []
dataF['mobilePhone'] = []
dataF['mobilePhoneModel'] = []
dataF['metrics'] = []
for row in data1:
    dataF['date'].append(str(row[0]))
    dataF['deviceCategory'].append(str(row[1]))
    dataF['mobilePhone'].append(str(row[2]))
    dataF['mobilePhoneModel'].append(str(row[3]))
    dataF['metrics'].append(str(row[4]))
print(dataF)
```

```python
df = pd.DataFrame.from_dict(dataF)
```
